2. Suponga que el valor de una acci贸n var铆a cada d铆a de 16 a 25 d贸lares con una distribuci贸n uniforme.
   * Calcule la probabilidad de que el valor de la acci贸n sea superior a 19 d贸lares. (**Rta**: $2/3$)
   * Calcule la probabilidad de que el valor de la acci贸n est茅 entre 19 y 22 d贸lares. (**Rta**: $1/3$)
   * Halle el cuartil superior (el $25 \%$ de todos los d铆as que la acci贸n est谩 por encima de 驴qu茅 valor?). (**Rta**: $22.75$)

```python
import scipy.stats as stats

# Definir los par谩metros de la distribuci贸n uniforme
a = 16
b = 25

# Calcular la probabilidad de que el valor de la acci贸n sea superior a 19 d贸lares
p1 = 1 - stats.uniform.cdf(19, loc=a, scale=b-a)
print("La probabilidad de que el valor de la acci贸n sea superior a $19 es {:.2f}".format(p1))

# Calcular la probabilidad de que el valor de la acci贸n est茅 entre 19 y 22 d贸lares
p2 = stats.uniform.cdf(22, loc=a, scale=b-a) - stats.uniform.cdf(19, loc=a, scale=b-a)
print("La probabilidad de que el valor de la acci贸n est茅 entre $19 y $22 es {:.2f}".format(p2))

# Calcular el cuartil superior
p3 = stats.uniform.ppf(0.75, loc=a, scale=b-a)
print("El cuartil superior es {:.2f}".format(p3))

# Calcular la probabilidad de que el valor de la acci贸n sea mayor de 21 d贸lares dado que es mayor de 18 d贸lares
p4 = (1 - stats.uniform.cdf(21, loc=a, scale=b-a)) / (1 - stats.uniform.cdf(18, loc=a, scale=b-a))
print("La probabilidad de que el valor de la acci贸n sea mayor de $21 dado que es mayor de $18 es {:.2f}".format(p4))
```
